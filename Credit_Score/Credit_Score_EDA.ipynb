```python
import glob
import pandas as pd
```

```python
# List all csv files
csv_files = glob.glob("*.{}".format("csv"))
csv_files
```

```python
df_csv_append = pd.DataFrame()

# append the CSV files
for file in csv_files:
    df = pd.read_csv(file)
    df_csv_append = df_csv_append.append(df, ignore_index=True)
```
